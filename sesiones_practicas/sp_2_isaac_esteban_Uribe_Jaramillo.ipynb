### <span style="color:#2F749F;"><strong>Ejercicio 1: C치lculo de medidas de tendencia central</strong></span>

A partir de una muestra simulada de ingresos, calcula la **media, mediana y moda**, y luego a interpretarlas.

1. Usa el n칰mero base `528310` (debes reemplazarlo con los 칰ltimos 6 d칤gitos de tu documento de identificaci칩n).  
2. Con este n칰mero genera una semilla (`np.random.seed`) para que los resultados sean reproducibles.  
3. Los ingresos se simulan con una distribuci칩n normal:  
   - Media te칩rica: `(base % 10) + 2`  
   - Desviaci칩n est치ndar: `2`  
   - Tama침o de muestra: `30`  
4. Los valores negativos se reemplazan por cero (pues los ingresos no pueden ser negativos).  
5. Grafica tus resultados (ver imagen de ejemplo)

<p align="center">
    <img src="imgs\sp_2_e1.png" alt="sp_2_e1" width="900"  height="400">
</p>


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import scipy
```

```python
#Definimos la semilla
seed = np.random.seed(240624)
#Creamos un array de tama침o 30
media_teorica = (240624 % 10) + 2
ingresos = np.random.normal(loc=media_teorica, scale=2, size=30).round(1)
ingresos[ingresos < 0] = 0
display(ingresos)
```

```python
# Medidas de tendencia central
media = sum(ingresos) / len(ingresos)
mediana = np.median(ingresos)
moda = stats.mode(ingresos, keepdims=True)[0][0]

# Mostrar resultados
print("Medidas de tendencia central:")
print(f"Media:   {media:.2f} ingresos")
print(f"Mediana: {mediana:.2f} ingresos")
print(f"Moda:    {moda:.2f} ingresos")

# Visualizaci칩n
plt.figure(figsize=(10, 5))
sns.histplot(ingresos, bins=10, kde=False, color="#2F749F")

# L칤neas de tendencia central
plt.axvline(media, color='red', linestyle='--', linewidth=2, label=f'Media: {media:.2f}')
plt.axvline(mediana, color='green', linestyle='-', linewidth=2, label=f'Mediana: {mediana:.2f}')
plt.axvline(moda, color='purple', linestyle=':', linewidth=2, label=f'Moda: {moda:.2f}')

# Est칠tica
plt.title("Distribuci칩n ingresos anuales", fontsize=14)
plt.xlabel("Ingreso anual", fontsize=12)
plt.ylabel("Frecuencia", fontsize=12)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# Generamos variables aletorias ingresos y gastos con la seed = 240624
ingresos = np.random.normal(loc=240624 % 10, scale=2, size=30).round(2)
gastos = np.random.normal(loc=240624 % 10, scale=2, size=30).round(2)
print("Ingresos:", ingresos)
print("Gastos:", gastos)
```

```python
#Calculamos los oeficientes de relacion de Pearson y Spearman
coef_pearson = np.corrcoef(ingresos, gastos)[0, 1]
coef_spearman, _ = scipy.stats.spearmanr(ingresos, gastos)
print("Coeficientes de correlaci칩n:")
print(f'Pearson: {coef_pearson:.2f}')
print(f'Spearman: {coef_spearman:.2f}')
```

### <span style="color:#2F749F;"><strong>Ejercicio 6: Chi-Cuadrado</strong></span>

En esta parte, analiza si existe **asociaci칩n estad칤stica** entre dos variables categ칩ricas simuladas a partir de tu n칰mero de c칠dula.

1. Usa los **6 칰ltimos d칤gitos** de tu c칠dula como base para generar una semilla reproducible.
2. Simula una muestra de **60 personas** con dos variables categ칩ricas:
   - `Nivel educativo`: B치sico, Medio, Superior.
   - `Acceso a internet`: S칤, No.
3. Crea una **tabla de contingencia** entre ambas variables.
4. Aplica la **prueba de Chi cuadrado de independencia**.
5. Visualiza la distribuci칩n conjunta con un mapa de calor (ver la imagen de con la informaci칩n de ejemplo).
6. Interpreta el resultado.

<p align="center">
    <img src="imgs\sp_2_e6.png" alt="sp_2_e6" width="900"  height="700">
</p>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

seed = 240624
np.random.seed(seed)

# Simulate data for 60 people
n_samples = 60
nivel_educativo = np.random.choice(['B치sico', 'Medio', 'Superior'], size=n_samples)
acceso_internet = np.random.choice(['S칤', 'No'], size=n_samples)

# Create a contingency table
contingency_table = pd.crosstab(nivel_educativo, acceso_internet)
print("Tabla de Contingencia:")
print(contingency_table)

# Perform Chi-Square test
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)

print("\nResultados de la prueba Chi Cuadrado:")
print(f"Estad칤stico Chi-Cuadrado: {chi2:.2f}")
print(f"Valor p: {p:.3f}")
print(f"Grados de libertad: {dof}")
# print("Frecuencias esperadas:\n", expected) # Optional: uncomment to see expected frequencies

# Visualize the joint distribution with a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(contingency_table, annot=True, fmt='d', cmap='Blues')
plt.title("Distribuci칩n Conjunta: Nivel Educativo vs. Acceso a Internet")
plt.xlabel("Acceso a Internet")
plt.ylabel("Nivel Educativo")
plt.show()

# Interpret the result (example interpretation based on p-value)
alpha = 0.05
```

Interpretaci칩n:
Dado que el valor p (0.768) es mayor que el nivel de significancia (0.05),
no tenemos suficiente evidencia para rechazar la hip칩tesis nula.
No existe evidencia significativa de asociaci칩n entre el nivel educativo y el acceso a internet en esta muestra.
