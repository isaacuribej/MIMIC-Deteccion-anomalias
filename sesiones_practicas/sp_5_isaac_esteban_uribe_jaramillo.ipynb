````python
# Hacemos todos los imports necesarios
import os
import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
````

````python
df = pd.DataFrame({
        'Nombre': ['Ana', 'Luis', 'Pedro', None, 'Marta', 'Luis', 'Sof칤a'],
        'Edad': [25, np.nan, 35, 29, -5, 25, None],
        'Ciudad': ['Bogot치', 'Medell칤n', None, 'Medell칤n', 'Cali', 'Bogot치', 'Cali'],
        'Ingreso': [3500, 4800, np.nan, 5200, 5100, np.nan, 4700],
        'FechaIngreso': ['2023-01-01', '2023-01-05', None, '2023-01-10', '2023-01-12', None, '2023-01-15']
    })
df
````

````python
#A los valores nulos los cambiamos por nulos tipo NaN pero para el formato de fechas
df['FechaIngreso'] = pd.to_datetime(df['FechaIngreso'], errors='coerce')

# Formatear en dd-mm-YYYY (como string para visualizarlo, pero sigue siendo datetime internamente)
df['FechaIngreso'] = df['FechaIngreso'].dt.strftime('%d-%m-%Y')

#A la columna "FeechaIngreso" los valores nulos los reemplazamos por el valor anterior o siguiente
df['FechaIngreso'] = df['FechaIngreso'].ffill().bfill()
df
````

````python
#A la columna "Ingreso" los valores nulos los reemplazamos por la media de la columna
df['Ingreso'].fillna(df['Ingreso'].mean(), inplace=True)
````

````python
# Hacemis imputacion de valores negativos y nulos en la columna "Edad" con la mediana
df['Edad'] = df['Edad'].apply(lambda x: df['Edad'].median() if (pd.isnull(x) or x < 0) else x)
````

````python
# A la columna "Ciudad" los valores nulos los reemplazamos por el valor anterior o siguiente
df['Ciudad'] = df['Ciudad'].ffill().bfill()
df
````

*Nota:*  Para la columna de nombbres que tengan valores nulos, los dejaremos intectos porque  no nos podemos sacar sus nombres de debajo de la manga, tampoco sabemos a quien pertenecen, y no hay mas considencias como en edades que nos permitan imputarlo, entonces dejaremos la variable intacta

## <span style="color:#2F749F;"><strong>Ejercicio 3. Correcci칩n de errores tipogr치ficos o de codificaci칩n</strong></span>

2.  En el mismo notebook anterior, para el nuevo dataframe df, estandariza la variable ciudad utilizando python. 

````python
df_p3 = pd.DataFrame({
        'Ciudad': ['bogota', 'Bogot치', 'BOGOTA', 'bogot치', 'bogata', 'B칩gota', 'BogoTa', 'Cali', 'cal칤', 'medell칤n', 'medellin']
    })
df_p3
````

````python
# Aplicamos la funcion lower para pasar todo a minusculas
df_p3['Ciudad'] = df_p3['Ciudad'].str.lower()
# Remplazamos tildes por letras sin tilde
df_p3['Ciudad'] = df_p3['Ciudad'].str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
# Con la funcion get_closest_matches de la libreria fuzzywuzzy podemos agrupar los nombres similares, en nuestro caso las palabras similares a nuestra lista de ciudades ['bogota', 'cali', 'medellin'] seran remplezadas por estas



df_p3
````

````python
from difflib import get_close_matches

# Lista base de nombres v치lidos en min칰sculas
ciudades_ref = ['bogota', 'medellin', 'cali']

# Funci칩n para reemplazar por la coincidencia m치s cercana
def corregir_ciudad(ciudad):
    match = get_close_matches(ciudad, ciudades_ref, n=1, cutoff=0.7)
    return match[0] if match else ciudad

# Aplicar correcci칩n
df_p3['Ciudad'] = df_p3['Ciudad'].apply(corregir_ciudad)

df_p3
````
